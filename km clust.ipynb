```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('agent.csv')
df.columns
```

```python
X = df.values
X.shape
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss, '-o', color='blue')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 0)
y = kmeans.fit_predict(X)
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], s = 50, c = 'green', label = 'Star Agents')
plt.scatter(X[y == 1, 0], X[y == 1, 1], s = 50, c = 'orange', label = 'Potential Agents')
plt.scatter(X[y == 2, 0], X[y == 2, 1], s = 50, c = 'red', label = 'Out of Game')
plt.title('Clusters of Agents')
plt.xlabel('Days since last sale')
plt.ylabel('Frequency of sale')
plt.legend()
plt.show()
```
