```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
df = pd.read_csv('agent.csv')
df.columns
```

```python
X = df.values
X.shape
```

```python
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'),color_threshold=1)
plt.title('Dendrogram')
plt.xlabel('Agent')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
h_clus = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y = h_clus.fit_predict(X)
```

```python
plt.scatter(X[y == 0, 0], X[y == 0, 1], s = 50, c = 'green', label = 'Star Agents')
plt.scatter(X[y == 1, 0], X[y == 1, 1], s = 50, c = 'orange', label = 'Potential Agents')
plt.scatter(X[y == 2, 0], X[y == 2, 1], s = 50, c = 'red', label = 'Out of Game')
plt.title('Clusters of Agents')
plt.xlabel('Days since last sale')
plt.ylabel('Frequency of sale')
plt.legend()
plt.show()
```
